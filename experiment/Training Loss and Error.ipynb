```python
import seaborn as sns
```

```python
with open("ap/train.txt", "r") as f:
    data_ap = f.readlines()
```

```python
with open("arow/train.txt", "r") as f:
    data_arow = f.readlines()
```

```python
with open("l2sgd/train.txt", "r") as f:
    data_l2sgd = f.readlines()
```

```python
with open("lbfgs/train.txt", "r") as f:
    data_lbfgs = f.readlines()
```

```python
with open("pa/train.txt", "r") as f:
    data_pa = f.readlines()
```

```python
def get_losses(data):
    losses = []
    errors = []
    for i in data:
        if i.startswith("Loss"):
            losses.append(i.split(":")[1].strip())
        if i.startswith("Error"):
            errors.append(i.split(":")[1].strip())
    losses = list(map(float, losses))
    errors = list(map(float, errors))
    return losses
```

```python
loss_ap = get_losses(data_pa)
loss_arow = get_losses(data_arow)
loss_l2sgd = get_losses(data_l2sgd)
loss_lbfgs = get_losses(data_lbfgs)
loss_pa = get_losses(data_pa)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x=range(len(loss_l2sgd)), y=loss_l2sgd)
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
```

```python
plt.figure(figsize=(10, 8))
plt.plot(range(len(loss_ap)), loss_ap, label="ap")
plt.plot(range(len(loss_arow)), loss_arow, label="arow")
plt.plot(range(len(loss_l2sgd)), loss_l2sgd, label="l2sgd")
plt.plot(range(len(loss_lbfgs)), loss_lbfgs, label="lbfgs")
plt.plot(range(len(loss_pa)), loss_pa, label="pa")
plt.title("Training Loss VS Iteration Number")
plt.legend(loc="best")
```
